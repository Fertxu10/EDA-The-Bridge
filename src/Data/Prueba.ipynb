```python
columnas_a_borrar_2 = [    
    "Corner_Kicks_Home", "Pass_Success_per_Home", "Dangerous_Attacks_Home", "Dangerous_Attacks_Away", "Attacks_Home", "Attacks_Away", 
    "Completed_Passes_Home", "Completed_Passes_Away", "home_team_yellow_card", "away_team_yellow_card", "Goal_Attempts_Home", "Goal_Attempts_Away"
]

df = df.drop(columns=columnas_a_borrar_2)
```

```python
columnas_rellenables = [
    "Red_Cards_Home", "Red_Cards_Away",
    "Yellow_Cards_Home", "Yellow_Cards_Away",
    "Fouls_Home", "Fouls_Away",
    "Offsides_Home", "Offsides_Away"
]

df[columnas_rellenables] = df[columnas_rellenables].fillna(0)

df[columnas_rellenables].isnull().sum()

```

```python
# Asegurar que las columnas de goles estén en formato entero

df["home_score"] = pd.to_numeric(df["home_score"]).astype(int)

df["away_score"] = pd.to_numeric(df["away_score"]).astype(int)
```

```python
df.iloc[61120:61125]
```

```python
# Eliminar filas donde home_score o away_score sean el string "-"

# Crear máscara booleana para esas filas
condicion = (df["home_score"] == "-") & (df["away_score"] == "-")

# Usar drop con los índices que cumplen la condición
df = df.drop(index=df[condicion].index)


```

```python
# Resultado del partido
df["resultado"] = df.apply(lambda row: "local" if row["home_score"] > row["away_score"] 
                           else "visitante" if row["home_score"] < row["away_score"] 
                           else "empate", axis=1)

# Diferencia de goles
df["diferencia_goles"] = df["home_score"] - df["away_score"]

# Cambiamos la columna por su valor absoluto para tener la diferencia siempre en positivo
df["diferencia_goles"] = df["diferencia_goles"].abs()

# ¿Ganó el equipo con más posesión?
df["local_mayor_posesion"] = df["Ball_Possession_Home"] > 50
df["local_gana"] = df["resultado"] == "local"
df
```

```python
def imputar_offsides(df, equipo_col, offside_col):
    df_resultado = df.copy()
    equipos = df[equipo_col].unique()

    for equipo in equipos:
        # Filtro por equipo en su rol (local o visitante)
        mask_equipo = df[equipo_col] == equipo

        # Calcular la media de offsides cuando hay datos
        media_equipo = df.loc[mask_equipo & df[offside_col].notnull(), offside_col].mean()

        # Rellenar nulos con esa media
        df_resultado.loc[mask_equipo & df[offside_col].isnull(), offside_col] = media_equipo

    return df_resultado
```

```python
def imputar_faltas(df, equipo_col, fouls_col):
    df_resultado = df.copy()
    equipos = df[equipo_col].unique()

    for equipo in equipos:
        # Filtro por equipo en su rol (local o visitante)
        mask_equipo = df[equipo_col] == equipo

        # Calcular la media de offsides cuando hay datos
        media_equipo = df.loc[mask_equipo & df[fouls_col].notnull(), fouls_col].mean()

        # Rellenar nulos con esa media
        df_resultado.loc[mask_equipo & df[fouls_col].isnull(), fouls_col] = media_equipo

    return df_resultado
```

```python
df = imputar_offsides(df, "home_team", "Offsides_Home")
```

```python
df = imputar_offsides(df, "away_team", "Offsides_Away")
```
